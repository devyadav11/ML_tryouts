```python
words = open('names.txt', 'r').read().splitlines()
print(words[:10])
```

```python
b = {}
for w in words:
  chs = ['<s>'] + list(w) + ['<e>']
  for ch1, ch2 in zip(chs, chs[1:]):
    bigram = (ch1, ch2)
    b[bigram] = b.get(bigram, 0) + 1

print(b)
```

```python
characters = sorted(list(set(''.join(words))))
print(characters)
```

```python
c2i = {c:i+1 for i,c in enumerate(characters)}
c2i['.'] = 0
i2c = {i:c for c,i in c2i.items()}      # It returns a view of the dictionaryâ€™s key-value pairs as tuples

print(c2i)
print(i2c)
```

```python
N = torch.zeros(27,27).float()
b = 0
for w in words:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = c2i[ch1]
    ix2 = c2i[ch2]
    N[ix1, ix2] += 1
    b += 1
print(N)
print(b)
```

```python
p = N/N.sum(1, keepdims=True)
print(p[0].sum())
```

```python
x = torch.Generator().manual_seed(12345)  # fixes the random numbers

for i in range(10):
  starting = 0
  predicted_word = []
  while True:
    p = N[starting].float()

    #p = torch.ones(27)
    #p = p/p.sum()
    starting = torch.multinomial(p, num_samples=1, replacement=True, generator=x).item()
    next_char = i2c[starting]
    if next_char == '.':
      break
    predicted_word.append(next_char)

  print(''.join(predicted_word))
```
